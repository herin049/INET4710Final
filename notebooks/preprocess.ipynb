```python
#Drop Bad fields
dataframe = dataframe.drop(columns=["bathroomsPartial", "mainLevelBathrooms", "laundryFeatures", 
                                    "fireplaces", "fireplaceFeatures", "commonWalls", "buildingArea", 
                                   "aboveGradeFinishedArea", "belowGradeFinishedArea", "parking", 
                                   "coveredSpaces", "openParkingSpaces", "carportSpaces", "otherParking",
                                   "accessibilityFeatures", "entryLevel", "entryLocation", "spaFeatures", 
                                   "patioAndPorchFeatures", "waterfrontFeatures", "frontageType", "frontageLength", 
                                   "woodedArea", "vegetation", "lotSizeDimensions", "otherStructures", "additionalParcelsDescription"
                                   ,"landLeaseAmount", "zoningDescription", "architecturalStyle", "constructionMaterials",
                                   "windowFeatures", "propertyCondition", "builderModel", "electric", "hasElectricOnProperty", 
                                   "gas", "waterSources", "utilities", "greenBuildingVerificationType", "greenEnergyEfficient", 
                                   "greenIndoorAirQuality", "greenSustainability", "greenWaterConservation", "numberOfUnitsInCommunity",
                                   "numberOfUnitsVacant", "storiesTotal", "buildingFeatures", "buildingName", "elementarySchool",
                                   "elementarySchoolDistrict", "middleOrJuniorSchool", "middleOrJuniorSchoolDistrict", "highSchool",
                                   "highSchoolDistrict", "securityFeatures", "communityFeatures", "listingId", "buildingAreaSource",
                                   "state", "topography", "hasAdditionalParcels", "parcelNumber", "zoning", "yearBuiltEffective",
                                   "sewer", "rooms", "otherFacts", "streetAddress", "fencing"])
```

```python
#Filter NaN values
filtered_df = dataframe
filtered_df['bedrooms'] = filtered_df['bedrooms'].fillna(0)
filtered_df['bathrooms'] = filtered_df['bathrooms'].fillna(0)
filtered_df['bathroomsFull'] = filtered_df['bathroomsFull'].fillna(0)
filtered_df['bathroomsHalf'] = filtered_df['bathroomsHalf'].fillna(0)
filtered_df['bathroomsThreeQuarter'] = filtered_df['bathroomsThreeQuarter'].fillna(0)
filtered_df['bathroomsOneQuarter'] = filtered_df['bathroomsOneQuarter'].fillna(0)
filtered_df['basement'] = filtered_df['basement'].fillna("None")
filtered_df['hasHeating'] = filtered_df['hasHeating'].fillna(False)
filtered_df['hasCooling'] = filtered_df['hasCooling'].fillna(False)
filtered_df['hasFireplace'] = filtered_df['hasFireplace'].fillna(False)
filtered_df['furnished'] = filtered_df['furnished'].fillna(False)
filtered_df = filtered_df.dropna(axis=0, subset=['livingArea'])
filtered_df['garageSpaces'] = filtered_df['garageSpaces'].fillna(0)
filtered_df['hasAttachedGarage'] = filtered_df['hasAttachedGarage'].fillna(False)
filtered_df['hasGarage'] = filtered_df['hasGarage'].fillna(False)
filtered_df['hasOpenParking'] = filtered_df['hasOpenParking'].fillna(False)
filtered_df['hasCarport'] = filtered_df['hasCarport'].fillna(False)
filtered_df['levels'] = filtered_df['levels'].fillna(0)
filtered_df['stories'] = filtered_df['stories'].fillna(0)
filtered_df['hasPrivatePool'] = filtered_df['hasPrivatePool'].fillna(False)
filtered_df['hasSpa'] = filtered_df['hasSpa'].fillna(False)
filtered_df['hasView'] = filtered_df['hasView'].fillna(False)
filtered_df['hasWaterfrontView'] = filtered_df['hasWaterfrontView'].fillna(False)
filtered_df['canRaiseHorses'] = filtered_df['canRaiseHorses'].fillna(False)
filtered_df['lotSize'] = filtered_df['lotSize'].fillna("0.00 acres")
filtered_df['hasAttachedProperty'] = filtered_df['hasAttachedProperty'].fillna(False)
filtered_df['hasLandLease'] = filtered_df['hasLandLease'].fillna(False)
filtered_df['homeType'] = filtered_df['homeType'].fillna("Other")
filtered_df['roofType'] = filtered_df['roofType'].fillna("Asphalt")
filtered_df['isNewConstruction'] = filtered_df['isNewConstruction'].fillna(False)
filtered_df['hasHomeWarranty'] = filtered_df['hasHomeWarranty'].fillna(False)
filtered_df['hasPetsAllowed'] = filtered_df['hasPetsAllowed'].fillna(True)
filtered_df['hasRentControl'] = filtered_df['hasRentControl'].fillna(False)
filtered_df['structureType'] = filtered_df['structureType'].fillna("Other")
filtered_df['isSeniorCommunity'] = filtered_df['isSeniorCommunity'].fillna(False)
filtered_df['cityRegion'] = filtered_df['cityRegion'].fillna("Other")
filtered_df = filtered_df.dropna(axis=0, subset=['zipcode', 'city', 'price'])
filtered_df['homeStatus'] = filtered_df['homeStatus'].fillna("FOR_SALE")
filtered_df['photoCount'] = filtered_df['photoCount'].fillna(0)
filtered_df['photoCount'] = filtered_df['imageIdCount'].fillna(0)
```

```python
import math

for index, row in filtered_df.iterrows():
    if(pd.isnull(row['flooring'])):
        filtered_df.at[index, 'flooring'] = "[]"
    if(pd.isnull(row['heating'])):
        filtered_df.at[index, 'heating'] = "[]"
    if(pd.isnull(row['cooling'])):
        filtered_df.at[index, 'cooling'] = "[]"
    if(pd.isnull(row['appliances'])):
        filtered_df.at[index, 'appliances'] = "[]"
    if(pd.isnull(row['parkingFeatures'])):
        filtered_df.at[index, 'parkingFeatures'] = "[]" 
    if(pd.isnull(row['exteriorFeatures'])):
        filtered_df.at[index, 'exteriorFeatures'] = "[]" 
    if(pd.isnull(row['foundationDetails'])):
        filtered_df.at[index, 'foundationDetails'] = "[]" 
    if(pd.isnull(row['exteriorFeatures'])):
        filtered_df.at[index, 'exteriorFeatures'] = "[]" 
    if(pd.isnull(row['view'])):
        filtered_df.at[index, 'view'] = "[]" 
```
